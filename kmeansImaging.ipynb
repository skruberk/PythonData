```python
#k means color quantization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.image as mpimg
```

```python
image = mpimg.imread('Brain_MRI.jpg')
```

```python
#image
```

```python
plt.figure(figsize=(4,4),dpi=100)
plt.imshow(image)
```

```python
image.shape
```

```python
(h,w,c) = image.shape #extract image dimensions
```

```python
image2 = image.reshape(h*w,c) #reshape to 2d
```

```python
model = KMeans(n_clusters=6) #quantize to six colors
```

```python
labels = model.fit_predict(image2) #predict labels for each pixel 
```

```python
#RGB color code
model.cluster_centers_
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int) #coordinates of RGB values of six colors
```

```python
quant_img = np.reshape(rgb_codes[labels], (h, w, c)) #map each pixel to nearest cluster cernter 
```

```python
plt.figure(figsize=(6,6),dpi=100)
plt.imshow(quant_img)
```

```python
model = KMeans(n_clusters=3) #quantize to 3 colors
```

```python
labels = model.fit_predict(image2)
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int) 
```

```python
quant_img = np.reshape(rgb_codes[labels], (h, w, c)) #map each pixel to nearest cluster cernter 
```
